```python
from bs4 import BeautifulSoup
import requests

def get_page(n):
    answer = requests.get(f'https://www.mos.ru/news/?page={n}')
    
    document = BeautifulSoup(answer.text, 'html.parser')
    
    links = document.find_all('div')
    for link in links:
        print(link)
        print()
        print()
    
```

```python
n_pages = 1

for i in range(n_pages):
    get_page(2)
```
