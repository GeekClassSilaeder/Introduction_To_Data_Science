```python
def preprocess(text):
    symb = ['\n', '"', ',', '.', '!', '&', '?', '/', '\\\\', ':', ';', '[', ']', '{', '}', '*', '@', '#', '$', '%', '^', '(', ')', '_', '-', '=', '>', '<']
    
    for s in symb:
        text = text.replace(s, '')
    
    text_arr = text.split(' ')
    for i in range(len(text_arr)):
        text_arr[i] = text_arr[i].lower()
    
    return text_arr
```

```python
import pandas as pd

def vectorize(texts, words):
    data = pd.DataFrame(columns=words)
    
    for text in texts:
        text_dict = dict()
        for word in text.split(' '):
            if word in text_dict.keys():
                text_dict[word] += 1
            else:
                text_dict[word] = 1
        data = pd.concat([data, pd.DataFrame([text_dict])])
        
    return data
```

```python
def bag_of_words(texts):
    s = set()
    for text in texts:
        for word in text.split():
            s.add(word)
            
    return list(s)
```
